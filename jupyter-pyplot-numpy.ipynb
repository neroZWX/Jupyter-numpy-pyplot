# 10-Use gradient descent

```python
def grad_m(x, y, m, c):
  return -2.0 * np.sum(x * (y - m * x - c))
def grad_c(x, y, m, c):
  return -2.0 * np.sum(y - m * x - c)
eta = 0.0001
m, c = 1.0, 1.0
change = True
L = np.array(PLength)#didnt finsh Q8,Q9
d = np.array(PWidth)#use PLength and PWidth instead of setosa
while change:
  mnew = m - eta * grad_m(L,d, m, c)
  cnew = c - eta * grad_c(L,d, m, c)
  if m == mnew and c == cnew:
    change = False
  else:
    m, c = mnew, cnew
print("m: %20.16f  c: %20.16f" % (m, c))
```

```python
pl.plot(L, m * L + c, 'b-')
pl.plot(L, d, 'k.')
pl.xlabel('petal_length')
pl.ylabel('petal_width')
pl.show()
```
